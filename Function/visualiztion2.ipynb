```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# box plot
x=np.array([1,4,1,5,2,8,7,9,4,5,2,5,1,6,7,5,4,6,1,5])
x.shape
```

```python
plt.boxplot(x, notch=True, vert=False)
plt.show()
```

```python
plt.hist(x, bins=range(0,x.max()+1,2), edgecolor='black')

plt.grid()
plt.xticks()
plt.yticks([])
plt.show()
```

```python
gap=range(0,x.max()+1,2)
plt.hist(x, bins=gap, edgecolor='black')

plt.grid()
plt.xticks(gap)
plt.yticks([])
plt.yticks(range(0,8+1,2))
plt.show()
```

```python
# scatter.plot
x=np.random.randn(100)
y=np.random.randn(100)
plt.scatter(x,y,s=10,alpha=.5,c='#90d')
plt.show()
```

```python
# scatter.plot
x=np.random.randn(100)
y=np.random.randn(100)
plt.scatter(x,y,s=100,alpha=.5,c=y,cmap='gist_rainbow')
plt.colorbar()
plt.show()
```
