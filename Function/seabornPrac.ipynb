```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib_inline
sns.set_theme(style="whitegrid")
```

```python

df=sns.load_dataset("penguins")
print(df)
```

```python
sns.displot(df, x="bill_length_mm", y="bill_depth_mm",col='sex',row='species')
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    hue="coherence", style="choice",
)
```

```python
print(dots)
```

```python
palette = sns.cubehelix_palette(light=.9, n_colors=8)
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    hue="coherence", style="choice", palette=palette,
)
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    size="coherence", style="choice",
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,xlim=(-100,600))
yticksR=list(range(0,int(titanic.age.max()+1),20))
plt.yticks(yticksR)
plt.show()

```

```python
sns.countplot(x='sex',data=titanic,hue='sex',palette="Paired")
sns.color_palette("Paired",8)
```

```python

sns.boxplot(x='class',y='age',hue='class',data=titanic,palette='rainbow',linewidth=2)
```

```python
sns.displot(titanic['fare'],kde=False,color='red',bins=30,legend=False)
```

```python
sns.swarmplot(data=titanic,x='class',y='age',hue='class',palette='rainbow')
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
g=sns.FacetGrid(col='sex',data=titanic,hue='sex',palette='dark').map(sns.distplot,'age',kde=False)
```
